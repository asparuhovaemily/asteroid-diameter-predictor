## Introduction

The primary motivation behind this project is to automate the process of asteroid measuring in order to address two main aspects: space exploration and scientific research as well as planetary defence and impact mitigation. Understanding the size and composition of asteroids is essential for future space exploration missions.

It is noted by NASA ([Asteroid Size Estimator, n.d.](#references)) that in order to determine the diameter of an asteroid, scientists must evaluate its albedo, meaning the amount of light it reflects back into space. In general, a larger asteroid is indicated by a greater albedo. However, an article ([Lea, 2023](#references)) states that the problem is that the asteroid's surface reflectivity affects the albedo. This indicates that measurements of an asteroid's size could be significantly unclear because a small, light-coloured asteroid could have a greater albedo than a larger, darker one.

Therefore accurate measurements of the diameter are crucial and not only for estimating the potential damage an asteroid could cause but also for determining the necessary measures to change its trajectory, if possible at all.

The primary goal of this project is to create a predictive model capable of accurately predicting the diameters of asteroids. This involves the application of machine learning algorithms to establish a reliable and precise prediction system. The successful implementation of this model aims to significantly enhance our understanding of the physical characteristics of asteroids, contributing valuable insights to the scientific community.

```python
sns.histplot(asteroids["diameter"], bins=50, kde=True, stat="density")
plt.xlabel("Diameter")
plt.ylabel("Density")
plt.title("Distribution of Asteroid Diameters")
plt.show()
```

When examining the histogram below it becomes evident that the majority of diameters fall withoin the range of -2 to 2 standard deviations from the mean. It is also visible that there is a tail that extends beyond 2 standard deviations from the mean and implies the presence of extreme outliers. However, eliminating all outliers might reduce variance and lead to a loss of valuable diversity in the dataset, which is crucial for robust model training. On the other hand, keeping all outliers may introduce bias into the model and analysis, potentially skewing the results.

```python
diameters = asteroids['diameter']
num_std_dev = 5

mean_diameter = diameters.mean()
std_dev = diameters.std()

x_values = np.arange(-num_std_dev, num_std_dev + 1)

bin_edges = [mean_diameter + i * std_dev for i in range(-num_std_dev, num_std_dev + 2)]
hist, bins = np.histogram(diameters, bins=bin_edges)

plt.figure(figsize=(10, 6))
plt.bar(x_values, hist, width=1, edgecolor='black')
plt.xlabel('Standard Deviations from Mean')
plt.ylabel('Frequency')
plt.title('Distribution of Asteroid Diameters in Standard Deviations from Mean')
plt.xticks(x_values, labels=x_values)
plt.show()
```

```python
diameters = asteroids['diameter']
mean_diameter = diameters.mean()
std_dev = diameters.std()

lower_bound = mean_diameter - 2 * std_dev
upper_bound = mean_diameter + 2 * std_dev

num_within_range = ((diameters >= lower_bound) & (diameters <= upper_bound)).sum()

percentage_within_range = (num_within_range / len(diameters)) * 100

print(f"The percentage of asteroids within the range of -2 to 2 standard deviations from the mean is: {percentage_within_range:.2f}%")
```

```python
std_dev = asteroids['diameter'].std()

threshold = 4

upper_bound = asteroids['diameter'].mean() + threshold * std_dev
lower_bound = asteroids['diameter'].mean() - threshold * std_dev

asteroids_filtered = asteroids[(asteroids['diameter'] >= lower_bound) & (asteroids['diameter'] <= upper_bound)]

num_outliers_removed = len(asteroids) - len(asteroids_filtered)
print("Number of outliers removed:", num_outliers_removed)

percentage_removed = (num_outliers_removed / len(asteroids)) * 100
print("Percentage of outliers removed: {:.2f}%".format(percentage_removed))
```

After a careful examination, I made the decision to remove any samples that extended more than **four** standard deviations from the mean. The choice was made in order to retain a robust model performance while also reducing the impact of outliers. Through experimentation, I observed that keeping fewer outliers led to an increase in the $R^{2}$, indicating improved model fit. However, it also resulted in an extremely high maximum error, suggesting potential instability in predictions. On the other hand, eliminating too many outliers resulted in a rather low maximum error but also in a decrease in the $R^{2}$. I tried to strike a balance between reducing the maximum error and maintaining the prediction accuracy of the model, and ultimately decided to eliminate outliers that were more than four standard deviations away.

```python
diameters = asteroids_filtered['diameter']
num_std_dev = 5

mean_diameter = diameters.mean()
std_dev = diameters.std()

x_values = np.arange(-num_std_dev, num_std_dev + 1)

bin_edges = [mean_diameter + i * std_dev for i in range(-num_std_dev, num_std_dev + 2)]
hist, bins = np.histogram(diameters, bins=bin_edges)

plt.figure(figsize=(10, 6))
plt.bar(x_values, hist, width=1, edgecolor='black')
plt.xlabel('Standard Deviations from Mean')
plt.ylabel('Frequency')
plt.title('Distribution of Asteroid Diameters in Standard Deviations from Mean (After abs(std) > 4 outlier removal)')
plt.xticks(x_values, labels=x_values)
plt.show()
```

While correlation heatmaps provide a visual representation of pairwise correlations between features, they might not capture multicollinearity that explicitly. This is why I chose to use the variance inflation factor (VIF), which assesses the correlation and strength of correlation among predictor variables in a regression model to see if after the removal of the highly correlated features there is still some problem with multicollinearity.

```python
asteroids_numeric_data_cleaned = asteroids_numeric_data.dropna()

vif_data = pd.DataFrame()
vif_data["feature"] = asteroids_numeric_data_cleaned.columns
vif_data["VIF"] = [variance_inflation_factor(asteroids_numeric_data_cleaned.values, i) for i in range(len(asteroids_numeric_data_cleaned.columns))]

vif_data = vif_data.sort_values(by='VIF', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(vif_data["feature"], vif_data["VIF"])
plt.xlabel('VIF')
plt.ylabel('Feature')
plt.title('Variance Inflation Factor (VIF)')
plt.grid(axis='x')
plt.show()
```

## Data Preparation

### Handling missing data

I will make a copy of the current `absolute_magnitude` column so after the imputation I can compare it with the result.

```python
original_absolute_magnitude = asteroids['absolute_magnitude'].copy()
```

I am going impute the values based on the asteroid's class. I will do this by using the minimum and maximum values for each class for generating random numbers within those ranges. With this method I am aiming to retain the overall distribution while also ensuring that the values that are going to be assigned are realistic.

I concluded it would be wiser to use a random number within that range rather than the mean value. The majority of missing values come from classes with absolute magnitudes that are extremely similar, as seen earlier in the data analysis. As a result, imputing data using the mean will give nearly identical values to each sample, disrupting the distribution and most likely having a negative impact on the predictions.

```python
min_absolute_magnitude_by_class = asteroids.groupby('class')['absolute_magnitude'].min()
max_absolute_magnitude_by_class = asteroids.groupby('class')['absolute_magnitude'].max()

def impute_absolute_magnitude(row):
    if pd.isnull(row['absolute_magnitude']):
        class_min = min_absolute_magnitude_by_class[row['class']]
        class_max = max_absolute_magnitude_by_class[row['class']]
        random_value = np.random.uniform(low=class_min, high=class_max)
        return random_value
    else:
        return row['absolute_magnitude']

asteroids['absolute_magnitude'] = asteroids.apply(impute_absolute_magnitude, axis=1)
```

By looking at the histograms below I can conclude that the original distribution has been preserved, indicating no disruption in the dataset.

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(original_absolute_magnitude, kde=True)
plt.title('Absolute Magnitude Before Imputation')
plt.xlabel('Absolute Magnitude')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(asteroids['absolute_magnitude'], kde=True)
plt.title('Absolute Magnitude After Imputation')
plt.xlabel('Absolute Magnitude')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

Based on my previous analysis I have decide that it would be better to drop the rows with missing `albedo` values. Since `albedo` doesn't correlate well enough with any of the other features then imputing the values would only introduce bias without any guarantee that the values are close to being realistic.
